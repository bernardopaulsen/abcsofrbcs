```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class OLG():

    def __init__(self,
                 which = 'Constant',
                 k = 0.01,              # capital
                 θ = 0.36,              # share of capital
                 κ = 4,                 # constant
                 γ = 0.6,
                 λ = 1):
        
        self.which = which
        self.k = k
        self.θ = θ
        self.κ = κ
        self.γ = γ
        self.λ = λ
        
    def next_k(self):
        
        ''' returns k_{t+1} given k_t '''
        
        k, θ, κ, λ  = self.k, self.θ, self.κ, self.λ
        
        if self.which == 'Constant':
                 
            return κ*k**θ
        
        if self.which == 'Stochastic':
        
            return κ*λ*k**θ
        
            
    def equilibrium_k(self):
        
        ''' returns the steady state capital '''
        
        θ, κ = self.θ, self.κ
        
        return κ**(1/(1-θ))
    
    def k_growth(self):
        
        ''' returns capital growth given the level of capital '''
        
        k, next_k = self.k, self.next_k()
        
        return next_k/k-1
    
    
    def next_λ(self):
        
        γ, λ = self.γ, self.λ
        
        ε = np.random.normal()*0.02
        
        return (1-γ) + γ*λ + ε
    
    def update(self):
        
        self.λ = self.next_λ()
        self.k = self.next_k()
    
    def figure_1(self):
        
        self.which = 'Constant'
        
        self.k = np.linspace(0,10,100)
        
        plt.plot(a.k, a.k)
        plt.plot(a.k, a.next_k())
        plt.xlabel('$k_{t}$')
        plt.ylabel('$k_{t+1}$')
        plt.show()
        
    def figure_2(self):
        
        self.which == 'Stochastic'
        
        self.k = 9
        
        lista = []
        
        for i in range(120):
            
            lista.append(self.k)
            self.update()
        
        plt.plot(lista)
        plt.show()
        
        
```
